```python
#CAESAR
def encrypt(key):
  text=input()
  for char in text:
    if char.isupper():
      print(chr((ord(char)+key-65)% 26+65),end="")
    elif char.islower():
      print(chr((ord(char)+key-97)%26+65),end="")
    else:
      print(char)
def decrypt(key):
  text=input()
  for char in text:
    if char.isupper():
      print(chr((ord(char)-key-65)% 26+65),end="")
    elif char.islower():
      print(chr((ord(char)-key-97)% 26+97),end="")
encrypt(3)
decrypt(3)
```

```python
import hashlib
with open("text.txt", "rb") as f:
    file_hash = hashlib.sha256(f.read()).hexdigest()
    print(file_hash)
trusted_hash =input()
if file_hash == trusted_hash:
  print("SAFE")
else:
  print("TAMPERED")
```

```python
#port scanning
import socket
target = "localhost"
s = socket.socket()
for port in range (1000):
  s.settimeout(1)
  result = s.connect_ex((target, port))
  if result == 0:
      print(f"Port {port} is OPEN")
  else:
      print(f"Port {port} is Closed")

s.close()
```

```python
def check_url(url):
    flags = []
    for word in ["login", "verify", "banking", "secure", "update"]:
        if word in url.lower():
            flags.append(f"Suspicious word: '{word}'")
    if len(url) > 75:
        flags.append("Too long")
    if "@" in url:
        flags.append("Contains '@'")
    print(f"\nURL: {url}")
    if flags:
        print(f" SUSPICIOUS: {flags}")
    else:
        print(" Clean")

# --- Try it out ---
check_url("https://definitely-not-a-suspicious-site.badsite.com/login")
check_url("https://google.com")
check_url("https://bank.com/login@fakebank.com/login")
```

```python
with open("ioc_list.txt", "w") as f:
    f.write("45.33.32.156\n198.51.100.23\n203.0.113.99\n")


with open("sample_traffic.log", "w") as f:
    f.write("2026-07-12 10:01:12 connection from 10.0.0.5\n")
    f.write("2026-07-13 10:02:44 connection from 45.33.32.156\n")
    f.write("2026-07-14 10:03:10 connection from 192.168.1.20\n")
    f.write("2026-07-14 10:04:55 connection from 203.0.113.99\n")


def load_iocs(filename):
    with open(filename) as f:
        return set(line.strip() for line in f if line.strip())

def scan_log(logfile, iocs):
    with open(logfile) as f:
        for line in f:
            for ip in iocs:
                if ip in line:
                    print("ALERT: Known malicious IP found ", line.strip())


iocs = load_iocs("ioc_list.txt")
print("Loaded", len(iocs), "IoCs")
scan_log("sample_traffic.log", iocs)
```

```python
import re
from collections import Counter

with open("login_attempts.log", "w") as f:
    f.write("2026-07-08 09:00:01 LOGIN SUCCESS user=alice ip=10.0.0.5\n")
    f.write("2026-07-08 09:01:15 LOGIN FAILED user=admin ip=203.0.113.99\n")
    f.write("2026-07-08 09:01:20 LOGIN FAILED user=admin ip=203.0.113.99\n")
    f.write("2026-07-08 09:01:25 LOGIN FAILED user=admin ip=203.0.113.99\n")
    f.write("2026-07-08 09:01:30 LOGIN FAILED user=admin ip=203.0.113.99\n")
    f.write("2026-07-08 09:01:35 LOGIN FAILED user=admin ip=203.0.113.99\n")
    f.write("2026-07-08 09:02:00 LOGIN SUCCESS user=bob ip=10.0.0.8\n")

failed_counter = Counter()

with open("login_attempts.log") as f:
    for line in f:
        if "LOGIN FAILED" in line:
            match = re.search(r"ip=(\S+)", line)
            if match:
                failed_counter[match.group(1)] += 1


print("Failed login attempts by IP:")
for ip, count in failed_counter.items():
    print(f" {ip}: {count} attempts")

    if count >= 5:
        print(f"ALERT: possible brute-force attack from {ip}")
```
